```python
import numpy as np #para manipular os 
from matplotlib import pyplot as plt #para plotar os gráficos
from sklearn.cluster import KMeans #para usar o KMeans
dataset = np.array(
[[-3, 4], 
[2, 0], 
[4, -4], 
[4, 3], 
[0, 2], 
[2, 1],
[1, -1], 
[-5, -3]])
kmeans = KMeans(n_clusters = 2, #numero de clusters
init = 'k-means++', n_init = 10, #algoritmo que define a posição dos clusters de maneira mais assertiva
max_iter = 300) #numero máximo de iterações
centroide= kmeans.fit_predict(dataset)
plt.scatter(dataset[:,1], dataset[:,0], c = centroide) #posicionamento dos eixos x e y
plt.xlim(-6, 6) #range do eixo x
plt.ylim(-6, 6) #range do eixo y
plt.grid() #função que desenha a grade no nosso gráfico
plt.scatter(kmeans.cluster_centers_[:,1],kmeans.cluster_centers_[:,0], s = 100, c = 'green') #posição de cada centroide no gráfico
plt.show()
```
